```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))

def gelu_derivative(x):
    a = 0.0356774 * x**3 + 0.797885 * x
    dg = 0.5 * (1 + np.tanh(a))
    return dg + 0.5 * x * (1 - np.tanh(a)**2) * (0.0535161 * x**2 + 0.797885)
```

```python
x = np.linspace(-4, 4, 1000)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(x, sigmoid(x), label="Sigmoid")
axs[0].plot(x, sigmoid_derivative(x), label="Sigmoid Derivative")
axs[0].set_title("Sigmoid & Derivative")
axs[0].legend()

axs[1].plot(x, relu(x), label="ReLU")
axs[1].plot(x, relu_derivative(x), label="ReLU Derivative")
axs[1].set_title("ReLU & Derivative")
axs[1].legend()

axs[2].plot(x, gelu(x), label="GELU")
axs[2].plot(x, gelu_derivative(x), label="GELU Derivative")
axs[2].set_title("GELU & Derivative")
axs[2].legend()

fig.suptitle('Activation Functions & Their Derivatives', fontsize=16)
# plt.show()
plt.savefig("activation_functions.png")

```
